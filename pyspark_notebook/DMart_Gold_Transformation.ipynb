#### Reading data from Sliver layer to Gold layer

```python
df_order = spark.read.parquet("Files/DMartSliver_Data/Sliver_Order_Data")
df_review =spark.read.parquet("Files/DMartSliver_review/Review")
df_socialmedia =spark.read.parquet("Files/DMartSliver_socialmedia/SocialMedia")
df_weblogs =spark.read.parquet("Files/DMartSliver_web_logs/WebLogs")
```

#### Aggregating the weblogs,social media and review to get some insights of customer experience on each product

##### Writing the data into Gold Layer lakehouse

```python
## Aggregation on the weblog data to see the per customer engagement for each page and action
df_weblogs =spark.read.parquet("Files/DMartSliver_web_logs/WebLogs")
df_weblogs = df_weblogs.groupBy("user_id","page","action").count()
df_weblogs.write.mode("overwrite").parquet("Files/DMartGold_weblogs/weblogs")
```

```python
## Aggregation on the socialmedia data the sentiment for each platform
df_socialmedia =spark.read.parquet("Files/DMartSliver_socialmedia/SocialMedia")
df_socialmedia = df_socialmedia.groupBy("platform","sentiment").count()
df_socialmedia.write.mode("overwrite").parquet("Files/DMartGold_socialmedia/socialmedia")
```

```python
## Aggregation on the review to get Average rating on the each prodcut
df_review =spark.read.parquet("Files/DMartSliver_review/Review")
df_review = df_review.groupBy("product_id").agg(round(avg("rating"),2).alias("AvgRating"))
df_review.write.mode("overwrite").parquet("Files/DMartGold_reviews/reviews")
```
